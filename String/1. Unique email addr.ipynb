```java
import java.util.HashSet;
import java.util.Set;

public class EmailValidator {
    public int numUniqueEmails(String[] emails) {
        Set<String> emailSet = new HashSet<>();

        for (String email : emails) {
            String[] arr = email.split("@");
            String localName = arr[0];
            String domain = arr[1];
            
            // Handle the '+' character rule
            String ignoredPlus = localName.split("\\+")[0];

            // Handle the '.' character rule
            StringBuilder sb = new StringBuilder();
            for (String str : ignoredPlus.split("\\.")) {
                sb.append(str);
            }

            // Reconstruct the normalized email and insert into Set
            emailSet.add(sb.append("@").append(domain).toString());
        }

        return emailSet.size();
    }
}
```

```java
// 1. Define your array of multiple inputs here
String[] inputEmails = {
    "test.email+alex@leetcode.com",
    "test.e.mail+bob.cathy@leetcode.com",
    "testemail+david@leetcode.com",
    "testemail.bob+cathy@leetcode.com"
};

// 2. Instantiate your class
EmailValidator validator = new EmailValidator();

// 3. Execute the method and print results
int uniqueCount = validator.numUniqueEmails(inputEmails);

System.out.println("Input Array: " + java.util.Arrays.toString(inputEmails));
System.out.println("Result     : " + uniqueCount + " unique email(s) found.");
```
