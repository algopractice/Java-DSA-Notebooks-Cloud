## Reason
The reason the first cell didn't print anything is a quirk of how Java Notebooks (JShell) work behind the scenes:
When you write a public class Test with a public static void main method inside a notebook cell, you are only defining the class. You haven't actually told Java to run the method yet. JShell simply saves the class layout into its memory (which is why it has a green checkmark showing it compiled successfully in 0.0s).
## How to execute it 
To make it print Hello, you have to explicitly call that main method you just created.Add a new line right below your class definition (or in the empty cell underneath it) to trigger it:

```java
Test.main(new String[] {});
```
